```python
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
```

```python
x= np.linspace(-3,3,100)
y= np.sqrt(9-x**2)
y1 = -np.sqrt(9-x**2)
plt.plot (x,y, c='k')
plt.plot (x,y1, c='k')
plt.axis('scaled')
x_s= np.linspace(-3,-3,100)
ys=x
plt.plot(x_s,ys, c='r')
plt.plot(-x_s,ys, c='r')
plt.plot(ys,x_s, c='r')
plt.plot(-x_s,ys, c='r')
plt.plot(ys,-x_s, c='r')
plt.savefig("circ.png")
```

```python
# plotting convergence

err= []
x=range(10,2000,10)
for i in x:
    ll= i
    x_n,y_n= np.random.uniform(-3,3,ll), np.random.uniform(-3,3,ll)
    x_in=x_n[np.where(x_n**2+y_n**2<9)]
    arr=len(x_in)/len(x_n) *36
    err.append(arr-np.pi*9)

    
```

```python
plt.plot(x,err)
plt.xlabel("total number of points")
plt.ylabel(" estimated area - real area ")
plt.savefig("plot.png")
```
